```python
from pathlib import Path
import pandas as pd

DATASET_PATH = Path('../data/steam_dataset_2.csv')
```

```python
steam_df = pd.read_csv(
    DATASET_PATH,
    delimiter=',',
    decimal='.',
    encoding='utf-8',
    low_memory=False)
```

```python
steam_df.columns
```

```python
years = ('2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024', '2025', '2026')

steam_df_copy = steam_df[steam_df['Release date'].str.endswith(years)].copy()
```

```python
columns_to_drop = ['Header image', 'Full audio languages', 
                   'Required age', 'Achievements', 'Notes',
                   'Website', 'Support url', 'Support email',
                   'Discount', 'DLC count', 'About the game']

steam_df_copy.drop(columns=columns_to_drop, inplace=True)
```

```python
steam_df_copy.columns
```

```python
steam_df_copy.to_csv('data/steam_dataset.csv', index=False)
```
